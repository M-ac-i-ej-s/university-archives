```python
import pandas
import matplotlib.pyplot as plt
```

```python
#a)
miasta = pandas.read_csv('miasta.csv', sep=',')
print(miasta.values)
```

```python
#b)
miasta.loc[len(miasta)] = [2010, 460, 555, 405]
print(miasta)
```

```python
#c)

plt.plot(miasta['Rok'],miasta['Gdansk'], 'ro-')
plt.xlabel('Lata')
plt.ylabel('Liczba ludnosci[w tys.]')
plt.show()
```

```python
#d)

plt.plot(miasta['Rok'],miasta['Gdansk'], label='Gdansk')
plt.plot(miasta['Rok'],miasta['Poznan'], label='Poznan')
plt.plot(miasta['Rok'],miasta['Szczecin'], label='Szczecin')
plt.xlabel('Lata')
plt.ylabel('Liczba ludnosci[w tys.]')
plt.legend()
plt.show()
```

```python
#e)

stdGdansk = (miasta['Gdansk'] - miasta['Gdansk'].mean()) / miasta['Gdansk'].std()
stdPoznan = (miasta['Poznan'] - miasta['Poznan'].mean()) / miasta['Poznan'].std()
stdSzczecin = (miasta['Szczecin'] - miasta['Szczecin'].mean()) / miasta['Szczecin'].std()

print(stdGdansk.mean(), stdGdansk.std())
print(stdPoznan.mean(), stdPoznan.std())
print(stdSzczecin.mean(), stdSzczecin.std())
```

```python
#f)

normGdansk = (miasta['Gdansk'] - miasta['Gdansk'].min()) / (miasta['Gdansk'].max() - miasta['Gdansk'].min())
normPoznan = (miasta['Poznan'] - miasta['Poznan'].min()) / (miasta['Poznan'].max() - miasta['Poznan'].min())
normSzczecin = (miasta['Szczecin'] - miasta['Szczecin'].min()) / (miasta['Szczecin'].max() - miasta['Szczecin'].min())

print(normGdansk.max(), normGdansk.min())
print(normPoznan.max(), normPoznan.min())
print(normSzczecin.max(), normSzczecin.min())
```
