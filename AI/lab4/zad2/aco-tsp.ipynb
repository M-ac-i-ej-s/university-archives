```python
import matplotlib.pyplot as plt
import random

from aco import AntColony


plt.style.use("dark_background")


COORDS = (
    (20, 52),
    (43, 50),
    (20, 84),
    (70, 65),
    (29, 90),
    (87, 83),
    (73, 23),
    (55, 20),
    (50, 50),
    (20, 20),
    (80, 80),
    (30, 34),
    (40, 40),
)


# def random_coord():
#     r = random.randint(0, len(COORDS))
#     return r


def plot_nodes(w=12, h=8):
    for x, y in COORDS:
        plt.plot(x, y, "g.", markersize=15)
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches([w, h])


def plot_all_edges():
    paths = ((a, b) for a in COORDS for b in COORDS)

    for a, b in paths:
        plt.plot((a[0], b[0]), (a[1], b[1]))


plot_nodes()

colony = AntColony(COORDS, ant_count=400, alpha=0.6, beta=0.9, 
                    pheromone_evaporation_rate=0.60, pheromone_constant=1000.0,
                    iterations=100)

optimal_nodes = colony.get_path()

for i in range(len(optimal_nodes) - 1):
    plt.plot(
        (optimal_nodes[i][0], optimal_nodes[i + 1][0]),
        (optimal_nodes[i][1], optimal_nodes[i + 1][1]),
    )


plt.show()
```

```python
# conclusion:
# after increasing the number of ants, the path is more accurate
# program runs slower if the alpha, beta value is too high
# interations seems to not have a big impact on the result
# after increasing the pheromone_evaporation_rate, the path is longer then before
```
